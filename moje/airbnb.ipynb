
# Plan Czyszczenia Zbioru Danych (Airbnb)

## Faza 1: Zrzucenie Balastu (Quick Wins)
W tej fazie pozbywamy się śmieci, które tylko zżerają pamięć i zaburzają statystyki.
- [ ] **Usunięcie bezużytecznych kolumn:** Wyrzucenie kolumny `Unnamed: 0` (pozostałość po starym indeksie).
- [ ] **Usunięcie klonów:** Sprawdzenie i usunięcie zduplikowanych wierszy w skali 1:1 (ta sama oferta skopiowana dwa razy).

## Faza 2: Transformacja Typów Danych (Rzutowanie)
W tej fazie zmuszamy Pandas do poprawnego rozumienia liczb i dat.
- [ ] **`price`:** Wycięcie symbolu waluty (`$`) i konwersja kolumny ze zmiennej tekstowej na liczbową (`float`).
- [ ] **`number_of_stays`:** Konwersja na liczby całkowite (`int`) – usunięcie wartości po przecinku.
- [ ] **`listing_added` & `last_review`:** Konwersja ze zwykłego tekstu na format daty (`datetime`).

## Faza 3: Inżynieria Cech (Rozbijanie tekstu)
W tej fazie wyciągamy użyteczne informacje z brudnych ciągów znaków.
- [ ] **`coordinates`:** Rozbicie nawiasu i podział na dwie osobne kolumny numeryczne: `latitude` i `longitude`.
- [ ] **`neighbourhood_full`:** Rozdzielenie tekstu po przecinku na dwie nowe kolumny: `city` oraz `neighbourhood`.
- [ ] **`room_type`:** Ujednolicenie kategorii (np. "private room" i "private" -> jedna nazwa).

## Faza 4: Obsługa Braków Danych (Missing Values)
W tej fazie łatamy dziury w bazie.
- [ ] **Weryfikacja 2075 braków:** Sprawdzenie, czy braki w ocenach wynikają z braku recenzji (`number_of_reviews == 0`).
- [ ] **Uzupełnianie luk:** Decyzja o uzupełnieniu lub usunięciu braków w `price` (238), `name` (5) i `host_name` (2).

## Faza 5: Walidacja Logiki Biznesowej (Sanity Checks)
W tej fazie sprawdzamy, czy dane nie łamią praw fizyki i logiki ekonomicznej.
- [ ] **Test kalendarza:** Sprawdzenie `availability_365` pod kątem wartości ujemnych lub > 365.
- [ ] **Test chronologii:** Sprawdzenie, czy `last_review` nie jest starsze niż `listing_added`.
- [ ] **Test skali ocen:** Sprawdzenie, czy `rating` i `5_stars` mieszczą się w poprawnym przedziale (np. 1-5).

## Faza 1 dropy


```python
airbnb = airbnb.drop(columns=['Unnamed: 0'] , errors='ignore')
airbnb = airbnb.drop_duplicates()
```

## Faza 2 transformacja typów

```python
airbnb['price'] = airbnb['price'].str.replace('$', '', regex=False).astype('float64')
airbnb['number_of_stays'] = airbnb['number_of_stays'].astype('Int64')
airbnb['listing_added'] = pd.to_datetime(airbnb['listing_added'])
airbnb['last_review'] = pd.to_datetime(airbnb['last_review'])
# print(airbnb[['price', 'number_of_stays', 'listing_added', 'last_review']].dtypes)
# airbnb
```

## Faza 3  rozbijanie i poprawianie  danych

```python
airbnb[['latitude', 'longitude']] = airbnb['coordinates'].str.strip('()').str.split(', ', expand=True).astype(float)
airbnb[['town','district']] = airbnb['neighbourhood_full'].str.split(', ' , expand=True)
airbnb['room_type'] = airbnb['room_type'].replace('private', 'Private room')
airbnb['room_type'] = airbnb['room_type'].replace('Private', 'Private room')
airbnb['room_type'] = airbnb['room_type'].replace('PRIVATE ROOM', 'Private room')
airbnb['room_type'] = airbnb['room_type'].replace('   Shared room      ', 'Shared room')
airbnb['room_type'] = airbnb['room_type'].replace('home', 'Entire home/apt')
airbnb = airbnb.drop(columns=['coordinates', 'neighbourhood_full'])
airbnb['room_type'].unique()
```

## Faza 4 obsługa Nan


```python
# airbnb[airbnb['name'].isna()]
airbnb = airbnb.dropna(subset=['name', 'host_name'])
mediana_dzielnicy = airbnb.groupby(['town' , 'district'])['price'].transform('median')
airbnb['price'] = airbnb['price'].fillna(mediana_dzielnicy)
mediana_dzielnicy = airbnb.groupby(['town'])['price'].transform('median')
airbnb['price'] = airbnb['price'].fillna(mediana_dzielnicy)
airbnb['visited'] = airbnb['last_review'] > airbnb['listing_added']
kolumny_numeryczne = ['reviews_per_month', 'rating', 'number_of_stays', '5_stars']
airbnb[kolumny_numeryczne] = airbnb[kolumny_numeryczne].fillna(0)
# airbnb[~airbnb.isna().any(axis=1)]
airbnb
```

## Faza 5 walidacja


```python
bledne_daty = airbnb[(airbnb['last_review'] < airbnb['listing_added'])]
print(f"Błędy w datach: {len(bledne_daty)}")
bledny_kalendarz = airbnb[(airbnb['availability_365'] < 0) | (airbnb['availability_365'] > 365)]
print(f"Błędy w dostępności: {len(bledny_kalendarz)}")
bledne_oceny = airbnb[(airbnb['rating']  < 0) | (airbnb['rating'] > 5)]
print(f"Błędy w ocenach (rating): {len(bledne_oceny)}")
bledne_5_stars = airbnb[(airbnb['5_stars'] < 0.0) | (airbnb['5_stars'] > 1.0)]
print(f"Błędy w odsetku 5 gwiazdek: {len(bledne_5_stars)}")
bledne_ceny = airbnb[(airbnb['price'] <= 0)]
print(f"Błędy w  cenach: {len(bledne_ceny)}")
bledne_ilosc_ocen = airbnb[(airbnb['number_of_reviews'] < 0)]
print(f"Błędy w  ilosci ocen: {len(bledne_ilosc_ocen)}")
```

```python
airbnb = airbnb.drop(index=bledne_daty.index)
airbnb = airbnb.drop(index=bledne_oceny.index)
airbnb = airbnb.drop(index=bledne_ceny.index)
# airbnb['last_review'] = airbnb['last_review'].fillna(pd.to_datetime('1970-01-01'))
airbnb.to_csv('airbnb_clean.csv', index=False)
```
